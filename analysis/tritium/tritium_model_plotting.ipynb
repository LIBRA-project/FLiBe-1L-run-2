```python
# ============================================================================
# DATA LOADING AND BASIC PLOTTING
# ============================================================================
from datetime import datetime
import json

# Create figure and axes (3 rows: cumulative OV, OV flow rate, temperature)
fig, (ax1, ax_flow, ax2) = plt.subplots(
    3,
    1,
    figsize=(8, 7),
    sharex=True,
    gridspec_kw={'height_ratios': [2, 0.8, 1]},
)

# Get cumulative activities and their errors
ov_cumulative = OV_stream_with_errors.get_cumulative_activity("total")
ov_errors = OV_stream_with_errors.get_cumulative_activity_errors("total")

# Convert times to numpy arrays for plotting
times_walls = np.array([t.magnitude for t in replacement_times_walls])

# Extract magnitudes for cumulative values (they have units)
ov_cumulative_mag = ov_cumulative.magnitude if hasattr(ov_cumulative, 'magnitude') else ov_cumulative

# Load timing data from general.json
with open("../../data/general.json", "r") as f:
    general_data_plot = json.load(f)

run_start = datetime.strptime(
    general_data_plot["timestamps"]["run_start"], "%m/%d/%Y %H:%M"
 )
flow_change_item = next(
    (
        item
        for item in general_data_plot["cover_gas"]["flow_rate_sample"]
        if item["value"] != 0
    ),
    None,
 )

if flow_change_item is not None:
    flow_change = datetime.strptime(flow_change_item["set_time"], "%m/%d/%Y %H:%M")
    flow_change_days = (flow_change - run_start).total_seconds() / 86400.0
else:
    flow_change_days = None

# Build step-wise OV flow-rate schedule from general.json
flow_rate_sample = general_data_plot["secondary_gas"]["flow_rate_sample"]
flow_rate_times = [
    (datetime.strptime(entry["set_time"], "%m/%d/%Y %H:%M") - run_start).total_seconds() / 86400.0
    for entry in flow_rate_sample
]
flow_rate_values = [entry["value"] for entry in flow_rate_sample]
flow_rate_times_step = flow_rate_times + [times_walls.max()]
flow_rate_values_step = flow_rate_values + [flow_rate_values[-1]]

# Build He + H2O intervals from secondary_gas.switched_to
switched_to = general_data_plot["secondary_gas"]["switched_to"]
switch_times_days = [
    (datetime.strptime(entry["gas_switch_time"], "%m/%d/%Y %H:%M") - run_start).total_seconds() / 86400.0
    for entry in switched_to
]
switch_types = [entry["type"] for entry in switched_to]
he_h2o_periods = []
for i, gas_type in enumerate(switch_types):
    if gas_type == "He + H2O":
        start_day = switch_times_days[i]
        end_day = switch_times_days[i + 1] if i + 1 < len(switch_times_days) else float(times_walls.max())
        he_h2o_periods.append((start_day, end_day))

# Mark selected flow changes only: mark only increases from 50 -> 250 SCCM
flow_change_mark_times = []
for i in range(1, len(flow_rate_sample)):
    prev_value = flow_rate_sample[i - 1]["value"]
    curr_value = flow_rate_sample[i]["value"]
    if prev_value == 50 and curr_value == 250:
        flow_change_mark_times.append(
            (datetime.strptime(flow_rate_sample[i]["set_time"], "%m/%d/%Y %H:%M") - run_start).total_seconds() / 86400.0
        )

# Get temperature data for bottom plot
temp_times = dip_tc.get_times_days()
temp_values = dip_tc.get_temperature()

# Find first heater setpoint increase to 750 C from general.json
dip_setpoints = general_data_plot['general_data']['temperature'].get('dip_setpoints', [])
heater_750_day = None
for j in range(1, len(dip_setpoints)):
    prev = dip_setpoints[j - 1]['value']
    curr = dip_setpoints[j]['value']
    if curr == 750 and curr > prev:
        heater_time = datetime.strptime(dip_setpoints[j]['set_time'], '%m/%d/%Y %H:%M')
        heater_750_day = (heater_time - run_start).total_seconds() / 86400.0
        break
if heater_750_day is None:
    for entry in dip_setpoints:
        if entry.get('value') == 750:
            heater_time = datetime.strptime(entry['set_time'], '%m/%d/%Y %H:%M')
            heater_750_day = (heater_time - run_start).total_seconds() / 86400.0
            break

# ============================================================================
# TUNABLE PLOT PARAMETERS - Adjust these values for fine-tuning
# ============================================================================
params = {
    # ---- Figure dimensions ----
    'figsize': (8, 5.8),
    
    # ---- Top plot: Tritium release ----
    'grid_alpha': 0.5,
    'line_width': 2,
    'marker_size': 5,
    'fill_alpha': 0.2,
    'irradiation_alpha': 0.3,
    
    # Top plot - OV (outer vessel) line
    'ov_color': "tab:green",
    'ov_label_x': 1.8,
    'ov_label_y': 2,
    'ov_label_text': 'Permeated tritium',
    'ov_label_fontsize': 14,
    
    # Top plot - Salt melting label
    'salt_melting_label_dx': 0.75,
    'salt_melting_label_y': 10,
    'salt_melting_label_fontsize': 12,
    
    # Top plot - Flow increase labels
    'flow_increase_label_dx': 0.75,
    'flow_increase_label_y_positions': [11.0, 7.0],
    'flow_increase_label_fontsize': 12,
    
    # Top plot - Axes labels and limits
    'ax1_ylabel': "Cumulative activity (Bq)",
    'ax1_ylabel_pad': 11,
    'ax1_title': "Permeated tritium release from FLiBe - Run 2",
    'ax1_title_pad': 20,
    'ax1_title_fontsize': 16,
    'ax1_ylim_min': 0,
    'ax1_ylim_max': 18,
    'ax1_xlim_min': -1,
    'ax1_xlim_max': 48,
    
    # ---- Bottom plot: Temperature ----
    'temp_color': "#C75702",
    'temp_linewidth': 1.5,
    'temp_alpha': 0.9,
    'temp_label_x': 40,
    'temp_label_y': 80,
    'temp_label_text': 'Salt temperature',
    'temp_label_fontsize': 12,
    
    # Bottom plot - "Neutrons ON" annotation
    'neutrons_text_x': 2,
    'neutrons_text_y': 6.8,
    'neutrons_text': 'Neutrons ON',
    'neutrons_text_color': 'red',
    'neutrons_text_alpha': 0.5,
    'neutrons_text_fontsize': 10,
    
    # Bottom plot - Arrow annotation
    'arrow_xy': (0.4, 4.5),
    'arrow_xytext': (3.8, 6.3),
    'arrow_color': 'red',
    'arrow_alpha': 0.3,
    'arrow_lw': 1.5,
    
    # Bottom plot - Axes labels and limits
    'ax2_xlabel': 'days',
    'ax2_xlabel_pad': 0,
    'ax2_ylabel': 'Temperature (°C)',
    'ax2_grid_alpha': 0.3,
    'ax2_ylim_min': 0,
    'ax2_ylim_max': 650,
}

# ============================================================================
# PLOTTING AND STYLING USING PARAMETERS
# ============================================================================

# Top plot: Tritium release (cumulative OV)
plt.sca(ax1)
ax1.grid(axis='y', alpha=params['grid_alpha'])

# Shade He + H2O intervals behind the data
for start_day, end_day in he_h2o_periods:
    ax1.axvspan(start_day, end_day, color='lightblue', alpha=0.25, zorder=0)

# Plot data lines with error bands
ax1.plot(times_walls, ov_cumulative_mag, color=params['ov_color'], linewidth=params['line_width'], marker='o', markersize=params['marker_size'], zorder=3)
ax1.fill_between(times_walls, ov_cumulative_mag - ov_errors, ov_cumulative_mag + ov_errors, color=params['ov_color'], alpha=params['fill_alpha'], zorder=2)

# Add selected flow-change markers
for mark_day in flow_change_mark_times:
    ax1.axvline(x=mark_day, color='tab:purple', linestyle='--', linewidth=1.2, alpha=0.65, zorder=1)

# Add labels to the first heater setpoint increase and flow-change markers
if heater_750_day is not None:
    for _ax in (ax1, ax_flow, ax2):
        _ax.axvline(x=heater_750_day, color='dimgray', linestyle=':', linewidth=1.2, alpha=1, zorder=6)
    ax1.text(
        0.2 + params['salt_melting_label_dx'],
        2 + params['salt_melting_label_y'],
        'Solid salt',
        color='gray',
        fontsize=params['salt_melting_label_fontsize'],
        rotation=90,
        va='center',
        ha='left',
        zorder=7,
    )
    ax1.text(
        heater_750_day + params['salt_melting_label_dx'],
        params['salt_melting_label_y'],
        'Molten salt',
        color='gray',
        fontsize=params['salt_melting_label_fontsize'],
        rotation=90,
        va='center',
        ha='left',
        zorder=7,
    )

for idx, mark_day in enumerate(flow_change_mark_times):
    label_y = params['flow_increase_label_y_positions'][idx] if idx < len(params['flow_increase_label_y_positions']) else params['flow_increase_label_y_positions'][-1]
    ax1.text(
        mark_day + params['flow_increase_label_dx'],
        label_y,
        'Flow rate increase',
        color='tab:purple',
        alpha=0.8,
        fontsize=params['flow_increase_label_fontsize'],
        rotation=90,
        va='center',
        ha='left',
        zorder=7,
    )

# Add data labels
ax1.text(params['ov_label_x'], params['ov_label_y'], params['ov_label_text'], 
         color=params['ov_color'], va='center', fontsize=params['ov_label_fontsize'], fontweight='bold', zorder=4)
ax1.text(1, 16.2, 'Moist He', color='#023047', fontsize=12, ha='left', va='bottom', zorder=4)
ax1.text(10, 16.2, 'Dry He', color='black', alpha=0.8, fontsize=12, ha='left', va='bottom', zorder=4)
ax1.text(30.5, 16.2, 'Moist He', color='#023047', fontsize=12, ha='left', va='bottom', zorder=4)

# Add irradiation shading
plot_irradiation(baby_model, facecolor="#EF5B5B", alpha=params['irradiation_alpha'])

# Set labels, title, and limits
ax1.set_ylabel(params['ax1_ylabel'], labelpad=params['ax1_ylabel_pad'], fontsize=10)
ax1.set_title(params['ax1_title'], loc="left", pad=params['ax1_title_pad'], fontsize=params['ax1_title_fontsize'])
ax1.set_ylim(params['ax1_ylim_min'], params['ax1_ylim_max'])
ax1.set_xlim(params['ax1_xlim_min'], params['ax1_xlim_max'])

# Middle plot: OV gas flow rate from general.json as a step plot
plt.sca(ax_flow)
ax_flow.grid(axis='y', alpha=0.2)
ax_flow.step(flow_rate_times_step, flow_rate_values_step, where='post', color='tab:purple', lw=1.5)
for start_day, end_day in he_h2o_periods:
    ax_flow.axvspan(start_day, end_day, color='lightblue', alpha=0.25, zorder=0)
# Add irradiation shading to middle plot
plot_irradiation(baby_model, facecolor="#EF5B5B", alpha=params['irradiation_alpha'])
ax_flow.text(1.7, 180, 'OV gas flow rate', color='tab:purple', fontsize=12, ha='left', va='bottom', fontweight='bold', zorder=6)
for mark_day in flow_change_mark_times:
    ax_flow.axvline(x=mark_day, color='tab:purple', linestyle='--', linewidth=1.2, alpha=0.65, zorder=3)
ax_flow.set_ylabel('Flow rate (SCCM)', fontsize=10)
ax_flow.set_xlim(params['ax1_xlim_min'], params['ax1_xlim_max'])
ax_flow.set_ylim(bottom=0, top=300)

# Bottom plot: Temperature profile
plt.sca(ax2)
ax2.plot(temp_times, temp_values, linewidth=params['temp_linewidth'], alpha=params['temp_alpha'], color=params['temp_color'])
ax2.axhline(y=459, color='tab:orange', linestyle=':', linewidth=1)
# Add melting point label
ax2.text(3, 490, r'FLiBe melting point', 
         color='tab:orange', fontsize=10)

# Add temperature label
ax2.text(params['temp_label_x'], params['temp_label_y'], params['temp_label_text'], 
         color=params['temp_color'], va='bottom', ha='center', fontsize=params['temp_label_fontsize'], fontweight='bold')

# Add "Neutrons ON" annotation with arrow
ax1.text(params['neutrons_text_x'], params['neutrons_text_y'], params['neutrons_text'], 
         color=params['neutrons_text_color'], alpha=params['neutrons_text_alpha'], fontsize=params['neutrons_text_fontsize'])
ax1.annotate('', xy=params['arrow_xy'], xytext=params['arrow_xytext'],
             arrowprops=dict(arrowstyle='->', color=params['arrow_color'], alpha=params['arrow_alpha'], lw=params['arrow_lw']))

# Add irradiation shading
plot_irradiation(baby_model, facecolor="#EF5B5B", alpha=params['irradiation_alpha'])

# Add power outage label
ax2.text(28.5, 280, r'Power outage', 
         color='gray', fontsize=10)
ax2.annotate('', xy=(25, 360), xytext=(28.3, 310),
             arrowprops=dict(arrowstyle='->', color='gray', alpha=0.8, lw=1))

# Set labels, grid, and limits
ax2.set_xlabel(params['ax2_xlabel'], labelpad=params['ax2_xlabel_pad'], fontsize=10)
ax2.set_ylabel(params['ax2_ylabel'])
ax2.grid(axis='y', alpha=params['ax2_grid_alpha'])
ax2.set_xlim(params['ax1_xlim_min'], params['ax1_xlim_max'])
ax2.set_ylim(params['ax2_ylim_min'], params['ax2_ylim_max'])

# Save and display figure
plt.tight_layout()
plt.savefig('outputs/OV_details.pdf', bbox_inches='tight')
plt.savefig('outputs/OV_details.svg', bbox_inches='tight')
plt.savefig('outputs/OV_details.png', bbox_inches='tight')
plt.show()
```

```python
# ============================================================================
# DATA LOADING AND BASIC PLOTTING
# ============================================================================
from datetime import datetime
import json

# Create figure and axes (4 rows: cumulative OV, OV flow rate, OV per-sample bars, temperature)
fig, (ax1, ax_flow, ax3, ax2) = plt.subplots(
    4,
    1,
    figsize=(8, 9),
    sharex=True,
    gridspec_kw={'height_ratios': [2, 0.8, 0.9, 1]},
)

# Get cumulative activities and their errors
ov_cumulative = OV_stream_with_errors.get_cumulative_activity("total")
ov_errors = OV_stream_with_errors.get_cumulative_activity_errors("total")

# Convert times to numpy arrays for plotting
times_walls = np.array([t.magnitude for t in replacement_times_walls])

# Extract magnitudes for cumulative values (they have units)
ov_cumulative_mag = ov_cumulative.magnitude if hasattr(ov_cumulative, 'magnitude') else ov_cumulative

# Load timing data from general.json
with open("../../data/general.json", "r") as f:
    general_data_plot = json.load(f)

run_start = datetime.strptime(
    general_data_plot["timestamps"]["run_start"], "%m/%d/%Y %H:%M"
 )
flow_change_item = next(
    (
        item
        for item in general_data_plot["cover_gas"]["flow_rate_sample"]
        if item["value"] != 0
    ),
    None,
)

if flow_change_item is not None:
    flow_change = datetime.strptime(flow_change_item["set_time"], "%m/%d/%Y %H:%M")
    flow_change_days = (flow_change - run_start).total_seconds() / 86400.0
else:
    flow_change_days = None

# Build step-wise OV flow-rate schedule from general.json
flow_rate_sample = general_data_plot["secondary_gas"]["flow_rate_sample"]
flow_rate_times = [
    (datetime.strptime(entry["set_time"], "%m/%d/%Y %H:%M") - run_start).total_seconds() / 86400.0
    for entry in flow_rate_sample
]
flow_rate_values = [entry["value"] for entry in flow_rate_sample]
flow_rate_times_step = flow_rate_times + [times_walls.max()]
flow_rate_values_step = flow_rate_values + [flow_rate_values[-1]]

# Build He + H2O intervals from secondary_gas.switched_to
switched_to = general_data_plot["secondary_gas"]["switched_to"]
switch_times_days = [
    (datetime.strptime(entry["gas_switch_time"], "%m/%d/%Y %H:%M") - run_start).total_seconds() / 86400.0
    for entry in switched_to
]
switch_types = [entry["type"] for entry in switched_to]
he_h2o_periods = []
for i, gas_type in enumerate(switch_types):
    if gas_type == "He + H2O":
        start_day = switch_times_days[i]
        end_day = switch_times_days[i + 1] if i + 1 < len(switch_times_days) else float(times_walls.max())
        he_h2o_periods.append((start_day, end_day))

# Mark selected flow changes only: mark only increases from 50 -> 250 SCCM
flow_change_mark_times = []
for i in range(1, len(flow_rate_sample)):
    prev_value = flow_rate_sample[i - 1]["value"]
    curr_value = flow_rate_sample[i]["value"]
    # Only mark increases from 50 to 250
    if prev_value == 50 and curr_value == 250:
        flow_change_mark_times.append(
            (datetime.strptime(flow_rate_sample[i]["set_time"], "%m/%d/%Y %H:%M") - run_start).total_seconds() / 86400.0
        )

# Get temperature data for bottom plot
temp_times = dip_tc.get_times_days()
temp_values = dip_tc.get_temperature()
# Find first heater setpoint increase to 750 C from general.json
dip_setpoints = general_data_plot['general_data']['temperature'].get('dip_setpoints', [])
heater_750_day = None
for j in range(1, len(dip_setpoints)):
    prev = dip_setpoints[j - 1]['value']
    curr = dip_setpoints[j]['value']
    if curr == 750 and curr > prev:
        heater_time = datetime.strptime(dip_setpoints[j]['set_time'], '%m/%d/%Y %H:%M')
        heater_750_day = (heater_time - run_start).total_seconds() / 86400.0
        break
if heater_750_day is None:
    for entry in dip_setpoints:
        if entry.get('value') == 750:
            heater_time = datetime.strptime(entry['set_time'], '%m/%d/%Y %H:%M')
            heater_750_day = (heater_time - run_start).total_seconds() / 86400.0
            break

# ============================================================================
# TUNABLE PLOT PARAMETERS - Adjust these values for fine-tuning
# ============================================================================
params = {
    # ---- Figure dimensions ----
    'figsize': (8, 5.8),
    
    # ---- Top plot: Tritium release ----
    'grid_alpha': 0.5,
    'line_width': 2,
    'marker_size': 5,
    'fill_alpha': 0.2,
    'irradiation_alpha': 0.3,
    
    # Top plot - OV (outer vessel) line
    'ov_color': "tab:green",
    'ov_label_x': 1.8,
    'ov_label_y': 2,
    'ov_label_text': 'Permeated tritium',
    'ov_label_fontsize': 14,
    
    # Top plot - Salt melting label
    'salt_melting_label_dx': 0.75,
    'salt_melting_label_y': 10,
    'salt_melting_label_fontsize': 12,
    
    # Top plot - Flow increase labels
    'flow_increase_label_dx': 0.75,
    'flow_increase_label_y_positions': [11.0, 7.0],
    'flow_increase_label_fontsize': 12,
    
    # Top plot - Axes labels and limits
    'ax1_ylabel': "Cumulative activity (Bq)",
    'ax1_ylabel_pad': 11,
    'ax1_title': "Permeated tritium release from FLiBe - Run 2",
    'ax1_title_pad': 20,
    'ax1_title_fontsize': 16,
    'ax1_ylim_min': 0,
    'ax1_ylim_max': 18,
    'ax1_xlim_min': -1,
    'ax1_xlim_max': 48,
    
    # ---- Bottom plot: Temperature ----
    'temp_color': "#C75702",
    'temp_linewidth': 1.5,
    'temp_alpha': 0.9,
    'temp_label_x': 40,
    'temp_label_y': 80,
    'temp_label_text': 'Salt temperature',
    'temp_label_fontsize': 12,
    
    # Bottom plot - "Neutrons ON" annotation
    'neutrons_text_x': 2,
    'neutrons_text_y': 6.8,
    'neutrons_text': 'Neutrons ON',
    'neutrons_text_color': 'red',
    'neutrons_text_alpha': 0.5,
    'neutrons_text_fontsize': 10,
    
    # Bottom plot - Arrow annotation
    'arrow_xy': (0.4, 4.5),
    'arrow_xytext': (3.8, 6.3),
    'arrow_color': 'red',
    'arrow_alpha': 0.3,
    'arrow_lw': 1.5,
    
    # Bottom plot - Axes labels and limits
    'ax2_xlabel': 'days',
    'ax2_xlabel_pad': 0,
    'ax2_ylabel': 'Temperature (°C)',
    'ax2_grid_alpha': 0.3,
    'ax2_ylim_min': 0,
    'ax2_ylim_max': 650,
}

# ============================================================================
# PLOTTING AND STYLING USING PARAMETERS
# ============================================================================

# Top plot: Tritium release (cumulative OV)
plt.sca(ax1)
ax1.grid(axis='y', alpha=params['grid_alpha'])

# Shade He + H2O intervals behind the data
for start_day, end_day in he_h2o_periods:
    ax1.axvspan(start_day, end_day, color='lightblue', alpha=0.25, zorder=0)

# Plot data lines with error bands
ax1.plot(times_walls, ov_cumulative_mag, color=params['ov_color'], linewidth=params['line_width'], marker='o', markersize=params['marker_size'], zorder=3)
ax1.fill_between(times_walls, ov_cumulative_mag - ov_errors, ov_cumulative_mag + ov_errors, color=params['ov_color'], alpha=params['fill_alpha'], zorder=2)

# Add selected flow-change markers
for mark_day in flow_change_mark_times:
    ax1.axvline(x=mark_day, color='tab:purple', linestyle='--', linewidth=1.2, alpha=0.65, zorder=1)

# Add labels to the first heater setpoint increase and flow-change markers
if heater_750_day is not None:
    for _ax in (ax1, ax_flow, ax3, ax2):
        _ax.axvline(x=heater_750_day, color='dimgray', linestyle=':', linewidth=1.2, alpha=1, zorder=6)
    ax1.text(
        0.2 + params['salt_melting_label_dx'],
        2 + params['salt_melting_label_y'],
        'Solid salt',
        color='gray',
        fontsize=params['salt_melting_label_fontsize'],
        rotation=90,
        va='center',
        ha='left',
        zorder=7,
    )
    ax1.text(
        heater_750_day + params['salt_melting_label_dx'],
        params['salt_melting_label_y'],
        'Molten salt',
        color='gray',
        fontsize=params['salt_melting_label_fontsize'],
        rotation=90,
        va='center',
        ha='left',
        zorder=7,
    )

for idx, mark_day in enumerate(flow_change_mark_times):
    label_y = params['flow_increase_label_y_positions'][idx] if idx < len(params['flow_increase_label_y_positions']) else params['flow_increase_label_y_positions'][-1]
    ax1.text(
        mark_day + params['flow_increase_label_dx'],
        label_y,
        'Flow rate increase',
        color='tab:purple',
        alpha=0.8,
        fontsize=params['flow_increase_label_fontsize'],
        rotation=90,
        va='center',
        ha='left',
        zorder=7,
    )

# Add data labels
ax1.text(params['ov_label_x'], params['ov_label_y'], params['ov_label_text'], 
         color=params['ov_color'], va='center', fontsize=params['ov_label_fontsize'], fontweight='bold', zorder=4)
ax1.text(1, 16.2, 'Moist He', color='#023047', fontsize=12, ha='left', va='bottom', zorder=4)
ax1.text(10, 16.2, 'Dry He', color='black', alpha=0.8, fontsize=12, ha='left', va='bottom', zorder=4)
ax1.text(30.5, 16.2, 'Moist He', color='#023047', fontsize=12, ha='left', va='bottom', zorder=4)

# Add irradiation shading
plot_irradiation(baby_model, facecolor="#EF5B5B", alpha=params['irradiation_alpha'])

# Set labels, title, and limits
ax1.set_ylabel(params['ax1_ylabel'], labelpad=params['ax1_ylabel_pad'], fontsize=10)
ax1.set_title(params['ax1_title'], loc="left", pad=params['ax1_title_pad'], fontsize=params['ax1_title_fontsize'])
ax1.set_ylim(params['ax1_ylim_min'], params['ax1_ylim_max'])
ax1.set_xlim(params['ax1_xlim_min'], params['ax1_xlim_max'])

# Mark heater 750°C (dark gray solid line) on all subplots if found
if heater_750_day is not None:
    for _ax in (ax1, ax_flow, ax3, ax2):
        _ax.axvline(x=heater_750_day, color='dimgray', linestyle=':', linewidth=1.2, alpha=1, zorder=6)
# Set labels, grid, and limits
ax2.set_xlabel(params['ax2_xlabel'], labelpad=params['ax2_xlabel_pad'])
ax2.set_ylabel(params['ax2_ylabel'])
ax2.grid(axis='y', alpha=params['ax2_grid_alpha'])
ax2.set_xlim(params['ax1_xlim_min'], params['ax1_xlim_max'])
ax2.set_ylim(params['ax2_ylim_min'], params['ax2_ylim_max'])
# Middle-top plot: OV flow rate from general.json as a step plot
plt.sca(ax_flow)
ax_flow.grid(axis='y', alpha=0.2)
ax_flow.step(flow_rate_times_step, flow_rate_values_step, where='post', color='tab:purple', lw=1.5)
for start_day, end_day in he_h2o_periods:
    ax_flow.axvspan(start_day, end_day, color='lightblue', alpha=0.25, zorder=0)
# Add irradiation shading to middle-top plot
plot_irradiation(baby_model, facecolor="#EF5B5B", alpha=params['irradiation_alpha'])
ax_flow.text(1.7, 180, 'OV gas flow rate', color='tab:purple', fontsize=12, ha='left', va='bottom', fontweight='bold', zorder=6)
for mark_day in flow_change_mark_times:
    ax_flow.axvline(x=mark_day, color='tab:purple', linestyle='--', linewidth=1.2, alpha=0.65, zorder=3)
ax_flow.set_ylabel('Flow rate (SCCM)', fontsize=10)
ax_flow.set_xlim(params['ax1_xlim_min'], params['ax1_xlim_max'])
ax_flow.set_ylim(bottom=0, top=300)

# Middle-bottom: Per-sample OV activity (bars) with x-limits matching top plot
plt.sca(ax3)
for start_day, end_day in he_h2o_periods:
    ax3.axvspan(start_day, end_day, color='lightblue', alpha=0.25, zorder=0)
plot_bars(OV_stream, index=[time.to(ureg.day) for time in replacement_times_walls], bar_width = 0.5)
# Add irradiation shading to middle-bottom plot
plot_irradiation(baby_model, facecolor="#EF5B5B", alpha=params['irradiation_alpha'])
# Add selected flow-change markers to middle-bottom plot (same as top/middle-top)
for mark_day in flow_change_mark_times:
    ax3.axvline(x=mark_day, color='tab:purple', linestyle='--', linewidth=1.2, alpha=0.65, zorder=3)

ax3.grid(axis='y', alpha=0.3)
ax3.set_ylim(bottom=0, top=5)
ax3.set_xlim(params['ax1_xlim_min'], params['ax1_xlim_max'])
ax3.set_ylabel('Sample activity (Bq)', fontsize=10, labelpad=16)
# MDA line (same as previous per-sample plot)
MDA = 0.071  # Calculated by hand for demo, TODO create code to calculate automatically
ax3.axhline(y=MDA, color = 'red', linestyle = '-.', linewidth=1)
ax3.annotate('MDA per vial', (4.8, 0.35), xytext=(0,0), textcoords='offset points', ha='center', va='bottom', color='red')

# Bottom plot: Temperature profile
plt.sca(ax2)
ax2.plot(temp_times, temp_values, linewidth=params['temp_linewidth'], alpha=params['temp_alpha'], color=params['temp_color'])
ax2.axhline(y=459, color='tab:orange', linestyle=':', linewidth=1)
# Add melting point label
ax2.text(3, 490, r'FLiBe melting point', 
         color='tab:orange', fontsize=10)

# Add temperature label
ax2.text(params['temp_label_x'], params['temp_label_y'], params['temp_label_text'], 
         color=params['temp_color'], va='bottom', ha='center', fontsize=params['temp_label_fontsize'], fontweight='bold')

# Add "Neutrons ON" annotation with arrow
ax1.text(params['neutrons_text_x'], params['neutrons_text_y'], params['neutrons_text'], 
         color=params['neutrons_text_color'], alpha=params['neutrons_text_alpha'], fontsize=params['neutrons_text_fontsize'])
ax1.annotate('', xy=params['arrow_xy'], xytext=params['arrow_xytext'],
             arrowprops=dict(arrowstyle='->', color=params['arrow_color'], alpha=params['arrow_alpha'], lw=params['arrow_lw']))

# Add irradiation shading
plot_irradiation(baby_model, facecolor="#EF5B5B", alpha=params['irradiation_alpha'])

# Add power outage label
ax2.text(28.5, 280, r'Power outage', 
         color='gray', fontsize=10)
ax2.annotate('', xy=(25, 360), xytext=(28.3, 310),
             arrowprops=dict(arrowstyle='->', color='gray', alpha=0.8, lw=1))

# Set labels, grid, and limits
ax2.set_xlabel(params['ax2_xlabel'], labelpad=params['ax2_xlabel_pad'], fontsize=10)
ax2.set_ylabel(params['ax2_ylabel'])
ax2.grid(axis='y', alpha=params['ax2_grid_alpha'])
ax2.set_xlim(params['ax1_xlim_min'], params['ax1_xlim_max'])
ax2.set_ylim(params['ax2_ylim_min'], params['ax2_ylim_max'])

# Overlay transparent legend PNG on third plot (draw last, above everything)
from mpl_toolkits.axes_grid1.inset_locator import inset_axes
legend_img = plt.imread('graphics/Bubbler color legend alt.png')
legend_ax = inset_axes(ax3, width='30%', height='50%', loc='upper left')
legend_ax.imshow(legend_img)
legend_ax.set_axis_off()
legend_ax.set_facecolor((0, 0, 0, 0))
legend_ax.set_zorder(100)

# Save and display figure
plt.tight_layout()
plt.savefig('outputs/OV_details.pdf', bbox_inches='tight')
plt.savefig('outputs/OV_details.svg', bbox_inches='tight')
plt.savefig('outputs/OV_details.png', bbox_inches='tight')
plt.show()
```
